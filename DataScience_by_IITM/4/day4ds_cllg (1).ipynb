```python
from googlesearch import search
title = 'chat gpt trending newss'
data = search(title,num_results=5)

all_data = []
for i in data:
    all_data.append(i+'\n')


f = open(f'{title}.txt',mode = 'a')
f.write('appending new lines================================= \n')
f.writelines(all_data)
f.close()
print('done')


#a mode is called append mode it adds new data in the existing file without deleting the previours data 











```

```python
#+ in plus mode we can read and write both in the filess 

f = open('day_4.txt','r+')
print(f.read())
f.write(' \n this is again a new line ')
f.close()

#r+ mode helps to read then write in the file it doesnt override the previous data 
```

```python
f = open('day_4.txt','w+')
print(f.read())
f.write(' \n this is again a new line ')
f.close()

#w+ it overides the previous data and wrtie the new data in the file
```

```python
f = open('day_4.txt','w+')
f.write(' \n this is again a new line ')
f.seek(0)
print(f.read())
f.close()
#we send the cursor  back to the zero so that we can read and write side by side
```
