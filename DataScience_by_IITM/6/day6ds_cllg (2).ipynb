```python
class Atm:
    def __init__(self):
        print('enter your request')
        self.pin = 1111
        self.balance = 2000

    def change_pin(self):
        self.pin = int(input('Enter your new pin : '))
        print('your pin has been changed')

    def show_balance(self):
        print(f''' Your current balance  is {self.balance}''')

    def atm_performance_status(self):
        print('ATM working Fine 100%')
```

```python
pnb  = Atm()
```

```python
pnb.show_balance()
```

```python
class Atm_1:
    def __init__(self):
        print('enter your request')
        self._pin = 1111    # _helps to make the variable protected
        self._balance = 2000

    def change_pin(self):
        self._pin = int(input('Enter your new pin : '))
        print('your pin has been changed')

    def show_balance(self):
        print(f''' Your current balance  is {self._balance}''')

    def atm_performance_status(self):
        print('ATM working Fine 100%')

```

```python
axis = Atm_1()
```

```python
class Atm_2:
    def __init__(self):
        print('enter your request')
        self.__pin = 1111      #double _  makes the variable private
        self.__balance = 2000

    def change_pin(self):
        self.__pin = int(input('Enter your new pin : '))
        print('your pin has been changed')

    def show_balance(self):
        print(f''' Your current balance  is {self.__balance}''')

    def atm_performance_status(self):
        print('ATM working Fine 100%')
```

```python
cc = Atm_2
#private variable can not be accessed drectly 
```
